```python
# Filter columns to be used
df = data[['nama pembeli', 'nama barang']]

# Mengubah DataFrame menjadi Series (Ini adalah langkah yang salah)
df = pd.DataFrame(df)

# Convert the dataset into a list of transactions
transactions = []
for row in df.itertuples(index=False):
    transaction = [item for item in row if item != '']
    transactions.append(transaction)

# Determine the minimum support threshold values to be tested
min_support_values = [0.01, 0.03, 0.05]

# Helper function to check if an itemset is closed
def is_closed(itemset, closed_itemsets):
    for other_itemset in closed_itemsets:
        if itemset.issubset(other_itemset) and itemset != other_itemset:
            return False
    return True

# Create a file to store the output
output_file = open('output.txt', 'w')

# Loop through the minimum support threshold values
for min_support in min_support_values:
    # Write Minimum Support to the file
    output_file.write(f"Minimum Support {min_support}:\n")
    closed_itemsets = []

    # Count item frequencies
    item_counts = {}
    for transaction in transactions:
        for item in transaction:
            if item in item_counts:
                item_counts[item] += 1
            else:
                item_counts[item] = 1

    # Find frequently occurring items (frequent items)
    frequent_items = {item for item, count in item_counts.items() if count >= len(transactions) * min_support}

    # CT-PRO Algorithm
    for item in frequent_items:
        # Initialize itemset as a set containing only the current item
        itemset = {item}
        # Calculate the support of the itemset
        itemset_support = sum(1 for transaction in transactions if itemset.issubset(set(transaction)))
        # Check if the itemset is closed
        if is_closed(itemset, closed_itemsets):
            # Add the closed itemset to the list
            closed_itemsets.append(itemset)

    # Save "Closed Itemsets" to the file
    output_file.write("\nClosed Itemsets:\n")
    for itemset in closed_itemsets:
        output_file.write("Itemset: {}, Support: {:.4f}\n".format(list(itemset), sum(1 for transaction in transactions if itemset.issubset(set(transaction))) / len(transactions)))

    # Tambahkan baris kosong antara setiap hasil Minimum Support
    output_file.write("\n")

# Tutup file output
output_file.close()
```
